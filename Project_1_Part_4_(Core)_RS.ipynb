```python
#importing the dataset post Part 2 modifications (NAs in Item_Weight and Outlet_Size are removed/imputed, modified Item_Fat_Content values to be Low Fat/Regular only)
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

food_df = pd.read_csv('/content/drive/MyDrive/Coding Dojo 2022/02 Week 2: Pandas/sales_predictions_after_week_2.csv')
food_df.head()
```

```python
#price vs total sales

plt.style.use('seaborn-bright')
reg = sns.regplot(data=food_df,x='Item_MRP',y='Item_Outlet_Sales', line_kws={'color':'black', 'ls': '--'})
xlabels = ['${:,.0f}'.format(x) for x in reg.get_xticks()]
ylabels = ['${:,.0f}K'.format(y) for y in reg.get_yticks()/1000]
reg.set_xticklabels(xlabels);
reg.set_yticklabels(ylabels);
reg.set_xlabel('Item Price',fontsize=12);
reg.set_ylabel('Total Item Sales',fontsize=12);
reg.set_title('Relationship between Price and Total Sales',c='b',fontsize=16);
```

```python
#sales by item type

it_sales = sns.barplot(data = food_df, x = 'Item_MRP', y = 'Item_Type',orient='h')
plt.ticklabel_format(style='plain', axis='x')
xlabels = ['${:,.0f}'.format(y) for y in it_sales.get_xticks()]
it_sales.set_xticklabels(xlabels);
it_sales.set_xlabel('Sales',fontsize=14);
it_sales.set_ylabel('Item_Category',fontsize=14);
it_sales.set_title('Average Item price by Category',fontsize=16);

```
