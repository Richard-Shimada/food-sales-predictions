```python
#boxplot - item weight/visibility/price/sales
food_df.head()
plt.title('Item Weight')
plt.boxplot(food_df['Item_Weight']);
plt.show()
plt.title('Item Visibility')
plt.boxplot(food_df['Item_Visibility']);
plt.show()
plt.title('Item MRP')
plt.boxplot(food_df['Item_MRP']);
plt.show()
plt.title('Item Outlet Sales')
plt.boxplot(food_df['Item_Outlet_Sales']);
```

1) Item visibility and Sales have a number of outliers above the "Maximum" value. Also, these values are skewed towards the Maximum (ie the top whisker is longer)

2) Item Weight and Item MRP have no outliers and are more evenly distributed. The IQR for these columns are bigger than Item Visibility/Sales

```python
#heatmap
corr = food_df.corr()
sns.heatmap(corr,cmap='Reds',annot=True)
```
